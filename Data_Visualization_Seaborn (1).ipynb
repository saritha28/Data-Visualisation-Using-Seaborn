```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#it is useful to count the no. of time each unique value is required
sns.countplot(x=tips.sex)
```

```python
tips.shape
```

```python
sns.distplot(x=tips['tip'],bins=5)
#to check the distribution of the data
```

```python
#kernel density estimation plot
#better alternative to histogram as you can see data skewness
sns.kdeplot(x=tips['tip'],shade=True)
```

```python
#shows the distribution of two parameter together
#distribution and correlation b/w the variables

sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```

```python
#runs regression lines with two parameters; kind='reg' will create a regression line;
#A regression line, also known as a "best-fit line" or "trend line"
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='reg')
```

```python
#shows the distribution of two parameters
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='kde',shade=True)
```

```python
#to find the datsets available in seaborn
sns.get_dataset_names()
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
iris.species.value_counts()
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_length,hue=iris.species)
```

```python
#visualizing the multi dimensional relationships among features
#pairplot are extensions to joint plots
#in seaborn the hue parameter determines which column in the data frame should be used for correlation

sns.pairplot(iris,height=2.5,hue="species")
```

```python
#bar lots
sns.barplot(x='sex',y='tip',data=tips,hue='smoker')
```

```python
iris.head()
```

```python
#swarm plot is used to understand the data distribution

sns.swarmplot(x='species',y='sepal_length',data=iris)
plt.title("Swarm Plot")
```

```python
#violin plot again shows the how data is distributed and helps in determing outlier
#hue helps us understand data in more detail

sns.violinplot(x='sex',y='tip',data=tips,hue='time')
```

```python
#Heatmap
#the heatmap is a way of represneting the data in a 2-dimensional form.
#the data values are represented as colors in the graph
#the goal of the heatmap is to provide a colored visual  summary of information.
```

```python
tips.corr()
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(tips.corr(),
           yticklabels=True,
           cbar=True,
           #cmap='coolwarm',
           cmap='YlGnBu',
           annot=True,
           linewidth=0,
           linecolor='grey')
```

```python
#to check the null values; here we can see the white vertical lines are null values
plt.figure(figsize=(12,5))
sns.heatmap(tips.isna(),
           yticklabels=True,
           cbar=True,
           #cmap='coolwarm',
           cmap='YlGnBu',
           annot=True,
           linewidth=0,
           linecolor='grey')
```

```python
#Box plot
sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
#Relation plot
#it uses extra keywords to use maximum datapoints in graph.

sns.relplot(x='total_bill',y='tip',col='time',
           hue='size',style='smoker',size='size',
           palette='YlGnBu',markers=["D","o"], sizes=(10,125),
           edgecolor=".2",linewidth=.5,alpha=.75,
            data=tips)
```
